```python
column = ['OP_UNIQUE_CARRIER', 'OP_CARRIER_AIRLINE_ID', 'OP_CARRIER', 'TAIL_NUM', 'OP_CARRIER_FL_NUM', 'DEST', 'ORIGIN', 'DEP_TIME', 'DEP_DEL15', 'DEP_TIME_BLK', 'ARR_TIME', 'ARR_DEL15', 'DIVERTED', 'DISTANCE' ]

df = data.drop(columns = column)
df.head()
```

```python
df.isnull().sum()
```

```python
array = df.values
X = array [:, 0:4]
Y = array [:, 4]
array
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
x_test
```

```python
from sklearn.naive_bayes import GaussianNB
modelnb = GaussianNB()
modelnb.fit(x_train, y_train)
```

```python
nbtrain = modelnb.fit(x_train, y_train)
nbtrain.class_count_
```

```python
y_pred = nbtrain.predict(x_test)
y_pred
```

```python
np.array(y_test)
```

```python
nbtrain.predict_proba(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
ac = accuracy_score(y_test, y_pred) 
cm = confusion_matrix(y_test, y_pred) 
ac 
```

```python
import seaborn as sns #library untuk membuat grafik dan statistik
cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], #mendefinisikan confusion matrix
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
```
